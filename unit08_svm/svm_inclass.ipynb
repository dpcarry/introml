## Problem 1.  Margin

For the points below with binary labels:
*  Create a scatter plot of the points with different markers for the two classes
*  Find the weight and bias of the classifier that separates the two classes
*  Compute the distance to the classifier boundary for the points
*  Find the margin

```python
# Data
X = np.array([[0.5, 0.5], [1, 0.5], [0.5, 1.75], [0.75, 2.75], [1.1, 2.2], [2, 1], [3, 1.5]])
y = np.array([1, 1, 1, 0, 0, 0, 0])

# Create scatter plot
plt.scatter(X[y == 1, 0], X[y == 1, 1], marker='o', label='Class 1')
plt.scatter(X[y == 0, 0], X[y == 0, 1], marker='x', label='Class 0')

# Find weight and bias (this is a simplified example, you'd typically use an SVM solver)
# We can visually see a separating line, let's approximate it:
w = np.array([-1, 1])  # Approximate weight vector
b = -0.5  # Approximate bias

# Compute the distance to the classifier boundary
distance = np.abs(np.dot(X, w) + b) / np.linalg.norm(w)

# Find the margin (minimum distance to the boundary)
margin = np.min(distance)

# Plot the separating hyperplane
x_line = np.linspace(0, 3.5, 10)
y_line = -(w[0] * x_line + b) / w[1]
plt.plot(x_line, y_line, label='Separating Hyperplane')

# Print results
print(f"Weight: {w}")
print(f"Bias: {b}")
print(f"Margin: {margin}")

plt.legend()
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Linearly Separable Data with Separating Hyperplane')
plt.show()
```
